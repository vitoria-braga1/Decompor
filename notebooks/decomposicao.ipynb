### INTRODUÇÃO 

 # Projeto De Analise  // Desempenho De Operadores _CallMeMaybe //

## Problema de negócios 
CallMeMaybe atua no setor de telefonia e deseja avaliar o desempenho de seus operadores.
    
O que vai ser analisado:
  Os funcionário **identificado operadores ineficientes** com os seguimos comportamento.
- Alto número de chamadas recebidas perdidas
- Tempo de espera elevado em chamadas recebidas
- Baixo volume de chamadas de saída, quando aplicável

## Visão dos Dados

### Dataset ("telecom_dataset_us.csv")

 **Descrição das colunas:**

 * "user_id": identificador do cliente
 * "date": data em que as estatísticas foram coletadas
 * "direction": direção da chamada (`in` para chamadas recebidas, `out` para chamadas realizadas)
 * "internal": indica se a chamada foi interna entre operadores do mesmo cliente
 * "operator_id": identificador do operador
 * "is_missed_call": indica se a chamada foi perdida
 * "calls_count": número de chamadas
 * "call_duration": duração da chamada sem considerar o tempo de espera
 * "total_call_duration": duração total da chamada, incluindo o tempo de espera

### Dataset ("telecom_clients_us.csv")
 * "user_id": identificador do cliente
 * "tariff_plan": plano tarifário contratado
 * "date_start": data de início do contrato do cliente
     
## Avaliação inicial
   Sera realizada uma análise da qualidade dos dados:
- Verificação de valores ausentes
- Identificação de registros sem operador associado
- Verificação de valores inconsistentes (durações negativas ou nulas)
- Análise dos tipos de dados das colunas

Esta etapa e fundamental para o dados ser confiavel

## Preparação e Limpeza dos Dados
   As principais atividades desta etapa incluem:
- Conversão da coluna `date` para o formato de data
- Remoção ou tratamento de valores ausentes
- Exclusão de registros sem `operator_id`
- Padronização de variáveis categóricas
    
### Criação de Variáveis
    Serão criadas as seguintes métricas:
- **Tempo de espera**: total_call_duration - call_duration
- **Chamadas recebidas**: chamadas com `direction = 'in'`
- **Chamadas de saída**: chamadas com `direction = 'out'`
- **Chamadas perdidas**: chamadas com `is_missed_call = True`

## EDA
   Serão analisados:
- Distribuição do número de chamadas por operador
- Percentual de chamadas perdidas
- Distribuição do tempo de espera
- Volume de chamadas de saída por operador

## Definição dos Critérios  
   Com base nas analise, serão definidos critérios objetivos para classificar operadores como ineficientes, tais como:
- Percentual de chamadas perdidas acima do percentil 75
- Tempo médio de espera acima da média geral
- Número de chamadas de saída abaixo da mediana (quando aplicável)

## Testes de Hipóteses Estatísticas
   Nesta etapa serão realizados testes estatísticos para validar se as diferenças observadas são significativas.
- Teste t de Student ou teste de Mann-Whitney (dependendo da distribuição dos dados)
- Nível de significância adotado: 5% (α = 0,05)
### As métricas analisadas incluem:   
- *Tempo de espera*
- *Percentual Chamadas perdidas*

## Interpretação dos Resultados
   Os resultados dos testes estatísticos e análise exploratória serão foco no impacto para o negócio, identificando:
- Principais fatores associados à ineficiência
- Quantidade de operadores ineficientes
- Possíveis causas operacionai

## Conclusões e Recomendações
   Com base nos resultados obtidos, serão apresentadas conclusões e recomendações práticas, como:
- Treinamentos direcionados para operadores com alto tempo de espera
- Ajustes na distribuição de chamadas
- Monitoramento contínuo dos indicadores de desempenho
